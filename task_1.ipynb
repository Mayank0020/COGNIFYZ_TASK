```python
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df1["Has Table booking"] = label_encoder.fit_transform(df["Has Table booking"])

df1["Has Online delivery"] = label_encoder.fit_transform(df["Has Online delivery"])

df1["Is delivering now"] = label_encoder.fit_transform(df["Is delivering now"])

df1["Switch to order menu"] = label_encoder.fit_transform(df["Switch to order menu"])
```

```python
df1["Rating text"].value_counts()
```

```python
rating_mapping = {
    'Excellent': 5,
    'Very Good': 4,
    'Good': 3,
    'Average': 2,
    'Poor': 1,
    'Not rated': 0
}

df1["Rating text"] = df1["Rating text"].map(rating_mapping)
```

```python
df1
```

```python
from sklearn.preprocessing import OrdinalEncoder

ordinal = OrdinalEncoder()

restaurant_ids = df1[['Restaurant ID']]

ordinal_encoded_ids = ordinal.fit_transform(restaurant_ids)

# Replace the original column with the encoded values
df1['Restaurant ID'] = ordinal_encoded_ids
```

```python
x = df1.drop(columns = "Aggregate rating", axis = 1)
y = df1["Aggregate rating"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state= 42)
```

```python
x_train
```

```python
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor


linear_model = LinearRegression()
tree_decision = DecisionTreeRegressor(random_state= 42)
```

```python
linear_model.fit(x_train, y_train)
```

```python
tree_decision.fit(x_train,y_train)
```

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
y_pred_linear = linear_model.predict(x_test)
y_pred_tree = tree_decision.predict(x_test)
```

```python
y_pred_linear
```

```python
y_pred_tree
```

```python

metrics = pd.DataFrame({'Model': ['Linear Regression','Decision Tree Regression'], 'Mean_squared_error': [mean_squared_error(y_test, y_pred_linear), mean_squared_error(y_test, y_pred_tree)], 'Accuracy' : [linear_model.score(x_test, y_test)*100, tree_decision.score(x_test, y_test)*100], 'Mean_squared_error': [mean_squared_error(y_test, y_pred_linear), mean_squared_error(y_test, y_pred_tree)], 
                        'R_squared' : [r2_score(y_test, y_pred_linear), r2_score(y_test, y_pred_tree)]})
metrics
```

```python
df1.corr()
```

```python
correlation_matrix = df1.corr()

target_correlation = correlation_matrix['Aggregate rating']

target_correlation
```

```python
thresholds = {
    'weak': 0.3,
    'moderate': 0.7
}

# Iterate over the target_correlation series
for feature, correlation in target_correlation.items():
    # Categorize the correlation based on the predefined thresholds
    if abs(correlation) < thresholds['weak']:
        category = 'weak'
    elif abs(correlation) < thresholds['moderate']:
        category = 'moderate'
    else:
        category = 'strong'
    
    # Print the result
    print(f"The correlation between target variable and {feature} is :- {category}")
```

## Now as you can see the independent variable that have a strong correlation with the dependent variable "Aggregate Rating" are:- 
* Restaurant Name
* Locality
* Switch to Order
* Rating text
